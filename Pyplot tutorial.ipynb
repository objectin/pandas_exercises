```python
x = np.linspace(0, 2*np.pi, 50)
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))
plt.show()
```

```python
x = np.linspace(0, 2*np.pi, 50)

plt.plot(x, np.sin(x))
plt.plot(x, 10*np.cos(2*x))
plt.show()
```

```python
x = np.linspace(0, 2*np.pi, 50)
ax1 = plt.gca()
ax2 = plt.gca().twinx()

ax1.plot(x, np.sin(x))
ax2.plot(x, 10*np.cos(2*x), color='y')

plt.show()
```

```python
x1 = np.linspace(0, 2*np.pi, 50)
y1 = np.sin(x1)

x2 = np.linspace(0.0, 2*np.pi, 50)
y2 = np.tan(x2)

# 1*2 구간으로 나누고 1번째에 그리기
plt.subplot(2, 1, 1)
plt.plot(x1, y1)

# 1*2 구간으로 2번째에 그리기
plt.subplot(2, 1, 2)
plt.plot(x2, y2)

plt.show()
```

```python
y1 = npr.normal(size=50).cumsum()
y2 = npr.normal(size=100).cumsum()
y3 = npr.normal(size=50).cumsum()
y4 = npr.normal(size=100).cumsum()
```

```python
plt.subplot(2, 2, 1)
plt.plot(y1)

plt.subplot(2, 2, 2)
plt.plot(y2)

plt.subplot(223)
plt.plot(y3)

plt.subplot(224)
plt.plot(y4)

plt.show()
```

```python
x = np.linspace(0, 2*np.pi, 50)
plt.plot(np.sin(x), label='sin')
plt.plot(np.cos(x), label='cos')
plt.legend()

plt.show()
```

```python
x = np.linspace(0, 2*np.pi, 50)

plt.plot(np.sin(x))
plt.plot(np.cos(x))
plt.legend(['sin', 'cos'])
plt.show()
```
